# Finding the shipping address that have the same billing address.

If the shipping and billing addresses are the same, then it will be a legitimate transaction (1).

```python

df['Address_match'] = (df['Shipping_address'] == df['Billing_address']).astype(int)
```

```python
df['Address_match'].value_counts()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='Address_match', palette='coolwarm')
plt.title('Difference between shipping and billing addresses')
plt.xlabel('Address_match')
plt.ylabel('Count')
legend_labels = ['Not same', 'Same']  
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in sns.color_palette('coolwarm', 2)]
plt.legend(handles, legend_labels, title='Address Match')
plt.show()
```

It can be clearly seen that 10017 addresses do not match the shipping and billing addresses, relating the same to fraud transaction.




# Identifying which payment method is used more?

```python

data = df
df = pd.DataFrame(data)
payment_data = df.groupby('Payment_method')['Transaction_amount'].sum().reset_index()
sns.set_theme()
colors = sns.color_palette('pastel')
color_map = dict(zip(payment_data['Payment_method'], colors))
plt.figure(figsize=(8, 10))
plt.pie(payment_data['Transaction_amount'], labels=payment_data['Payment_method'], autopct='%1.1f%%', startangle=140, colors=[color_map[method] for method in payment_data['Payment_method']])
plt.title('Transaction Amount Distribution by Payment Method')
plt.axis('equal') 
plt.show()


```

The data reveals that the transaction amounts are fairly evenly distributed across four payment methods.From this distribution, we can conclude that there is no significant dominance of any single payment method, indicating a balanced usage of debit cards, credit cards, PayPal, and bank transfers among the transactions.

# Discovering which category of products  people buy the most?

```python
category_counts = df['Product_category'].value_counts().sort_values(ascending=True)
sorted_df = pd.DataFrame({'Product_category': category_counts.index, 'Count': category_counts.values})
plt.figure(figsize=(10, 6))
sns.barplot(x='Product_category', y='Count', data=sorted_df, palette='Blues')
plt.title('Distribution of Product Category', fontsize=15)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

 # Finding the age group who made the largest purchases?

```python
from matplotlib.ticker import FuncFormatter
df_clean = df.dropna(subset=['Customer_age', 'Transaction_amount'])
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']
df_clean['AgeCategory'] = pd.cut(df_clean['Customer_age'], bins=age_bins, labels=age_labels, right=False)
def rupee_format(x, pos):
    return '{:,.0f}'.format(x)
plt.figure(figsize=(12, 6))
sns.barplot(x='AgeCategory', y='Transaction_amount', data=df_clean, estimator=sum, errorbar=None, palette='Oranges')
plt.xlabel('Age Category', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.title('Total Revenue by Age Category', fontsize=15)
plt.gca().yaxis.set_major_formatter(FuncFormatter(rupee_format))
plt.show()
```

##
The age categories 26-35 and 36-45 are the highest revenue contributors. Both categories show significantly higher total revenue compared to the other age groups.The age category 46-55 also contributes a notable amount of revenue, though it's considerably lower than the top two categories.The age categories 19-25 and 0-18 contribute less to the total revenue compared to the top three categories

# Recognizing which payment method was most often utilized by which age group?

```python

df_clean = df.dropna(subset=['Customer_age', 'Transaction_amount', 'Payment_method'])
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']
df_clean['AgeCategory'] = pd.cut(df_clean['Customer_age'], bins=age_bins, labels=age_labels, right=False)
def rupee_format(x, pos):
    return '{:,.0f}'.format(x)
plt.figure(figsize=(12, 6))
sns.barplot(x='AgeCategory', y='Transaction_amount', hue='Payment_method', data=df_clean, estimator=sum, errorbar=None, palette='Reds')
plt.xlabel('Age Category', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.title('Total Revenue by Age Category and Payment Method', fontsize=15)
plt.gca().yaxis.set_major_formatter(FuncFormatter(rupee_format))
plt.legend(title='Payment Method')
plt.show()
```

#The graph illustrates that the age groups 26-35 and 36-45 are the primary drivers of total sales, with a strong preference for credit cards and PayPal. Younger and older age groups contribute less to the total revenue.








# Discovering devices that took longer transaction hour?

```python
from matplotlib.ticker import FuncFormatter
data = df
df_clean = pd.DataFrame(data)
hour_bins = [0, 4, 8, 12, 16, 20, 24]
hour_labels = ['0-4', '4-8', '8-12', '12-16', '16-20', '20-24']
df_clean['HourCategory'] = pd.cut(df_clean['Transaction_hour'], bins=hour_bins, labels=hour_labels, right=False)
def rupee_format(x, pos):
    return '{:,.0f}'.format(x)
plt.figure(figsize=(12, 6))
sns.barplot(x='HourCategory', y='Transaction_amount', hue='Device_used', data=df_clean, estimator=sum, ci=None, palette='Greens')
plt.xlabel('Hour Category', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.title('Total Revenue by Hour Category and Device Used', fontsize=15)
plt.gca().yaxis.set_major_formatter(FuncFormatter(rupee_format))
plt.legend(title='Device Used')
plt.show()
```

The highest total revenue is observed in the hour category 0-4, indicating that a significant number of transactions happen during these hours.
The revenue generated from 8-12, 12-16, 16-20 and 20-24 hour categories is very similar, indicating steady sales throughout the daytime hours.Across all time intervals, mobile devices generally generate slightly more revenue.



# Detecting months that witnessed the highest incidence of fraud cases?

```python
df['Transaction_date']=pd.to_datetime(df['Transaction_date'])
```

```python
df['month']=df['Transaction_date'].dt.month
```

```python
df['month'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = df
df = pd.DataFrame(data)
df = df.sort_values(by='month')
plt.figure(figsize=(10, 6))
sns.countplot(x='month', hue='Is_fraudulent', data=df, palette='Set2')
plt.title('Count of Fraudulent Transactions by month')
plt.xlabel('Transaction months')
plt.ylabel('Count')
plt.show()
```

The graph illustrates the comparison between non-fraudulent (0) and fraudulent (1) transactions across four different months.The count of non-fraudulent transactions (green bars) is significantly higher than the count of fraudulent transactions (orange bars) for all four months.In the first three months, the total number of transactions (both fraudulent and non-fraudulent) is relatively stable, with counts exceeding 30,000 each month.
There is a noticeable drop in the fourth month, where the total transaction count significantly decreases.

```python
value_counts=df.groupby(['month','Is_fraudulent']).size().reset_index(name='COLUMNS')
print(value_counts)
```

```python
pivot_table=value_counts.pivot(index='month',columns='Is_fraudulent',values='COLUMNS')
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table,annot=True,fmt='d',cmap='Set1')
plt.xlabel('Is_fraudulent')
plt.ylabel('month')
plt.show()


```

# Identifying the amount which is typically involved in most scams?

```python
fraud_data = df.groupby('Is_fraudulent')['Transaction_amount'].sum()
labels = ['Non-Fraudulent', 'Fraudulent']
plt.figure(figsize=(8, 8))
plt.pie(fraud_data, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Transaction Amount Distribution: Fraudulent vs Non-Fraudulent')
plt.axis('equal')  
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
fraud_data = df.groupby('Is_fraudulent')['Transaction_amount'].sum().reset_index()
fraud_data['Is_fraudulent'] = fraud_data['Is_fraudulent'].map({0: 'Non-Fraudulent', 1: 'Fraudulent'})
def rupee_format(x, pos):
    return f'{x:,.0f}'
formatter = FuncFormatter(rupee_format)
plt.figure(figsize=(10, 6))
sns.barplot(x='Is_fraudulent', y='Transaction_amount', data=fraud_data,palette='Purples')
plt.title('Transaction Amount Distribution: Fraudulent vs Non-Fraudulent')
plt.xlabel('Transaction Type')
plt.ylabel('Total Transaction Amount')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

Non-Fraudulent Transactions:The total transaction amount for non-fraudulent transactions is significantly higher compared to fraudulent transactions.
The bar for non-fraudulent transactions reaches close to 20,000,000.
Fraudulent Transactions:The total transaction amount for fraudulent transactions is considerably lower.
The bar for fraudulent transactions is much shorter, indicating a total transaction amount of less than 5,000,000.
